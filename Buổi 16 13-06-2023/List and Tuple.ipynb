### Thực hành: Thêm học sinh vào lớp

```python
danh_sach_lop = ['Đức', 'Hiếu', 'Hoàng', 'Bình', 'Nam']
hoc_sinh_moi = ['Hương', 'Linh']
# cách 1:
danh_sach_lop += hoc_sinh_moi
print(danh_sach_lop)
# cách 2:
danh_sach_lop.extend(hoc_sinh_moi)

# cách 3: 
danh_sach_lop.append('Allen')
print(danh_sach_lop)
```

### Xóa phần tử khỏi list

```python
danh_sach_lop = ['Đức', 'Hiếu', 'Hoàng', 'Bình', 'Nam']
# cách 1: remove
danh_sach_lop.remove('Hiếu')
print(danh_sach_lop)
#cách 2: del
del danh_sach_lop[1]
print(danh_sach_lop)
# cách 3: pop
danh_sach_lop.pop()
print(danh_sach_lop)
print(danh_sach_lop)
```

### Sắp xếp list

```python
a = [1, 200, 43, 244, 56]
print(f'Danh sách trước khi sắp xếp: {a}')
# cách 1: sort()
a.sort()
print(f'Danh sách sau khi sắp xếp đúng chiều: {a}')
a.sort(reverse = 1)
print(f'Danh sách sau khi sắp xếp ngược chiều: {a}')
```

```python
a = [10, 34, 5, 67, 98, 45, 20]
b = sorted(a)
c = sorted(a, reverse = 1)
print(a)
print(b)
print(c)
```
