```python
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
    def __str__(self):
        return f'Name: {self.name}, quantity: {self.quantity}'
```

```python
from tkinter import *
from tkinter import messagebox

window = Tk()
window.title('Ingredient Management System')

ingredient_list = []

def add_ingredient():
    name_value = (name.get()).lower()
    quantity_value = quantity.get()
    check = False
    if name_value == '' or quantity_value == '':
        messagebox.showerror('Ingredient', 'Please enter name and quantity')
    
    for i in range (len(ingredient_list)):
        if ingredient_list[i].name == name_value:
            ingredient_list[i].quantity += int(quantity_value)
            messagebox.showinfo('Added successfully', f'Added {quantity_value} {name_value}')
            check = True
            break
            
    if check == False:    
        new_ingredient = Ingredient(name_value, int(quantity_value))
        ingredient_list.append(new_ingredient)
        
        print([ingredient.name for ingredient in ingredient_list])
        messagebox.showinfo('Added successfully', f'Added {quantity_value} {name_value}')

def delete_ingredient():
    delete_name = (search_name.get()).lower()
    check = False
    
    for i in range(len(ingredient_list)):
        if ingredient_list[i].name == delete_name:
            del ingredient_list[i]
            print([ingredient.name for ingredient in ingredient_list])
            messagebox.showinfo('Delete successfully', f'Deleted {delete_name} successfully')
            check = True
            break
    
    if check == False:
        messagebox.showerror('Cannot Delete', 'Cannot find the ingredient to delete')

def search_ingredient():
    searched_name = (search_name.get()).lower()
    check = False
    
    for i in range(len(ingredient_list)):
        if ingredient_list[i].name == searched_name:
            messagebox.showinfo('Ingredient Found', f'Ingredient Status: {ingredient_list[i]}')
            check = True
            break
    if check == False:
        messagebox.showerror('Ingredient Not Found', 'Cannot find the ingredient')

def reset():
    name.delete(0, 'end')
    quantity.delete(0, 'end')
    search_name.delete(0, 'end')

def take_ingredient():
    taken_ingredient = (name.get()).lower()
    taken_quantity = quantity.get()
    
    check = False
    if taken_ingredient == '' or taken_quantity == '':
        messagebox.showerror('Ingredient', 'Please enter name and quantity')
    
    for i in range (len(ingredient_list)):
        if ingredient_list[i].name == taken_ingredient:
            if ingredient_list[i].quantity >= int(taken_quantity):
                ingredient_list[i].quantity -= int(taken_quantity)
                messagebox.showinfo('Take successfully', f'Taken {taken_quantity} {taken_ingredient}')
            else:
                messagebox.showerror('Not enough ingredient', f'There are only {ingredient_list[i].quantity} of {taken_ingredient} left')
            check = True
            break
    if check == False:
            messagebox.showerror('Cannot Find Ingredient', 'There are no ingredient in store')
lbl_name = Label(window, text = 'Name:')
lbl_quantity = Label(window, text = 'Quantity:')

name = Entry(window)
quantity = Entry(window)
search_name = Entry(window)

btn_add = Button(window, text ='Add', command = add_ingredient)
btn_delete = Button(window, text = 'Delete', command = delete_ingredient)
btn_reset = Button(window, text = 'Reset', command = reset)
btn_search = Button(window, text = 'Search', command = search_ingredient)
btn_take = Button(window, text = 'Take', command = take_ingredient)

lbl_name.grid(row = 0, column = 0)
name.grid(row = 0, column = 1)
lbl_quantity.grid(row = 1, column =0)
quantity.grid(row=1,column=1)
btn_add.grid(row =2, column=0)
btn_delete.grid(row=2, column=1)
btn_reset.grid(row=2, column=2)
btn_take.grid(row =2, column = 3)
search_name.grid(row=3, column =0)
btn_search.grid(row=3, column =1)

window.mainloop()
```
